```python
#sparsing amenities column

import pandas as pd
import json


final_df['amenities'] = final_df['amenities'].astype(str)
# Convert single quotes to double quotes in the 'amenities' column
final_df['amenities'] =final_df['amenities'].str.replace("'", '"')

# Parse JSON strings in 'amenities' column
final_df['amenities'] = final_df['amenities'].apply(lambda x: json.loads(x))

# Expand the JSON into separate columns for each amenity
amenities_df = pd.json_normalize(final_df['amenities'])
print(amenities_df.head())

```

```python
#abbreviating amenities column
SC: Servant's Room or Servant's Quarter. This term indicates the presence of a room or space designated for household help or a servant.
GP: Garden/Park. It could indicate the availability of a garden or a park area within the property or its vicinity.
CPA: Car Parking Area. This might represent the availability of dedicated space for parking cars.
RWH: Rainwater Harvesting. This feature implies a system or method designed to collect and store rainwater for later use, typically for irrigation, landscaping, or other purposes.
STP: Sewage Treatment Plant. This refers to a facility or mechanism for treating sewage or wastewater before releasing it back into the environment, ensuring cleaner disposal.
HK: Housekeeping Services. It may indicate the availability or provision of housekeeping or cleaning services within the property.
PB: Power Backup. This feature signifies a backup power supply or system (such as generators or UPS) to provide electricity during power outages.
VP: Visitor Parking. This amenity denotes designated parking spaces specifically allocated for visitors or guests.
```

```python
final_df = pd.concat([final_df, amenities_df], axis=1)

# Drop the original 'amenities' column if it's no longer needed
final_df.drop('amenities', axis=1, inplace=True)

# Now 'final_df' contains separate columns for each amenity
print(final_df.head())
```

```python
duplicate_columns={'LIFT','GYM', "POOL"}
final_df.drop(columns=duplicate_columns,inplace=True)
final_df.shape
```
